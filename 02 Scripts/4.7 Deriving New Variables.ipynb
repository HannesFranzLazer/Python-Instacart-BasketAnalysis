### 03. If-Statements with For-Loops

```python
# counting frequency of orders by day
df_ords_prods_merged_new['order_dow'].value_counts(dropna=False)
```

```python
# classifying days of the weeks by order frequency
result = []
for value in df_ords_prods_merged_new["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Finding the two busiest and two slowest days
day_counts = df_ords_prods_merged_new['order_dow'].value_counts()
busiest_days = day_counts.nlargest(2).index.tolist()
slowest_days = day_counts.nsmallest(2).index.tolist()
```

```python
# Define a function to apply to the 'order_dow' column
def label_day(day):
    if day in busiest_days:
        return 'Busiest days'
    elif day in slowest_days:
        return 'Slowest days'
    else:
        return 'Regular days'
```

```python
# Create the new column
df_ords_prods_merged_new['Busiest_days'] = df_ords_prods_merged_new['order_dow'].apply(label_day)
```

```python
print(df_ords_prods_merged_new['Busiest_days'].value_counts())
```

```python
df_ords_prods_merged_new.head(20)
```

Answer: Saturdays and Sundays are the busiest days, while Tuesdays and Wednesdays are the slowest days. A list was created finding the strongest and weakest order days of the week. Then a if-function was set up using the criteria we established in the syntax finding the busiest/least busiest days.

##### Task 4:

```python
# counting frequency of orders by hour of day
df_ords_prods_merged_new['order_hour_of_day'].value_counts(dropna=False)
```

```python
# defining hour_counts & hour in df
hour_counts = df_ords_prods_merged_new['order_hour_of_day'].value_counts(dropna=False)
hours = df_ords_prods_merged_new['order_hour_of_day'].unique()
```

```python
# Defining the busiest hours as top 25 percentile, average as in 25 to 75 percentile and slowest hours as bottom 25 percentile
busiest_hours = hour_counts[hour_counts > hour_counts.quantile(0.75)].index
average_hours = hour_counts[(hour_counts <= hour_counts.quantile(0.75)) & (hour_counts > hour_counts.quantile(0.25))].index
slowest_hours = hour_counts[hour_counts <= hour_counts.quantile(0.25)].index
```

```python
# applying if-function to create new variable/column
def label_period(hour):
    if hour in busiest_hours:
        return 'Most orders'
    elif hour in average_hours:
        return 'Average orders'
    else:
        return 'Fewest orders'
```

```python
# Apply the function to the 'order_hour_of_day' column
df_ords_prods_merged_new['busiest_period_of_day'] = df_ords_prods_merged_new['order_hour_of_day'].apply(label_period)
```

#### Task 5:
