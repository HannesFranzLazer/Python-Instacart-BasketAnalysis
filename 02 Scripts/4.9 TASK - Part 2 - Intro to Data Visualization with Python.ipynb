## 02. Data Visualization

### Bar Charts

```python
# creating a bar chart
df_user_history['order_dow'].value_counts().plot.bar()
```

```python
# bar chart with sorted observations
bar = df_user_history['order_dow'].value_counts().sort_index().plot.bar()
```

```python
# printing frequencies
df_user_history['order_dow'].value_counts()
```

```python
df_user_history['order_dow'].value_counts().sort_index()
```

```python
# using color in bar chart
df_user_history['order_dow'].value_counts().plot.bar(color=['purple', 'red', 'green', 'yellow', 'blue', 'orange', 'brown'])
```

```python
# exporting bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

### Creating Histograms and Scatterplots

```python
# creating a histogram
df_user_history['prices'].plot.hist(bins = 25)
```

```python
df_user_history['prices'].describe()
```

```python
# creating a scatterplot
sns.scatterplot(x = 'prices', y = 'prices',data = df_user_history)
```

```python
# listing products under 100 $
df_user_history.loc[df_user_history['prices'] > 100]
```

```python
# marking prices > 100$ as missing values in order to proceed with the scatterplot
df_user_history.loc[df_user_history['prices'] >100, 'prices'] = np.nan
```

```python
# checking whether the replacement was successful by running a max-value check on the “prices” column
df_user_history['prices'].max()
```

```python
# creating a histogram, considering only prices < 25$
hist = df_user_history['prices'].plot.hist(bins = 25)
```

```python
hist_2 = df_user_history['prices'].plot.hist(bins = 70)
```

```python
# exporting histogram
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```
